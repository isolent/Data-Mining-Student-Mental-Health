```python
q_vecs=[]
for c in tqdm_notebook(q_cols):
    vectorizer = TfidfVectorizer(ngram_range=(1,3), max_df=1000, min_df=1, stop_words=sw)
    vecs = vectorizer.fit_transform(ds[c])
    q_vecs.append(vecs)
```

# Dimensionalityreduction

```python
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
```

```python
qvecs_pca=[]
for qv in tqdm_notebook(q_vecs):
    pca=PCA(n_components=2)
    qv_transformed=pca.fit_transform(qv.toarray())
    qvecs_pca.append(qv_transformed)
```

```python
from matplotlib import pyplot as plt
```

```python
i=0
for v in qvecs_pca:
    x = v[:,0]
    y = v[:,1]
    plt.figure(figsize=(20,20))
    plt.scatter(x, y,s=ds.Grading*3)
    plt.title(q_cols[i])
    for j, txt in enumerate(ds.Name):
        plt.annotate(txt, (x[j], y[j]))
    
    
    i+=1
    plt.show()
```

# Whose answers are duplicates?

```python
from scipy.spatial import distance
```

```python
def get_double_answers(vecs, ans, dist_threshold=5):
    #vecs = q_vecs[7]
    #ans = ds.q8
    for i in range(0,vecs.shape[0]):
        #print()
        for j in range(i+1,vecs.shape[0]):
            #print(i,j, end='\r')
            d=distance.cityblock(vecs[i].toarray(),vecs[j].toarray())
            #print(i,j,round(d),'|', end='\r')
            if d<=dist_threshold:
                if ans[i]!=' ' and ans[j]!=' ':
                    print(ds.Name[i]+':', ans[i])
                    print()
                    print(ds.Name[j]+':', ans[j])
                    print('===================')
```

```python
get_double_answers(q_vecs[0], ds.q1)
```
